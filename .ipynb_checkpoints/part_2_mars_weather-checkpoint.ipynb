### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
html = browser.html
Soup = soup(html, 'html.parser')
```

```python
# Find the table
table = Soup.find('table')
```

```python
# Check if the table element exists
if table:
    # Initialize lists to store table data
    headers = []
    data = []

    # Extract table headers (usually found in the <th> elements)
    header_row = table.find('tr')
    header_cells = header_row.find_all('th')
    for cell in header_cells:
        headers.append(cell.text.strip())

    # Extract table data from rows and cells
    data_rows = table.find_all('tr')[1:]  # Skip the first row with headers

    for row in data_rows:
        row_data = []
        cells = row.find_all('td')  # You can also use 'th' for table header cells
        for cell in cells:
            row_data.append(cell.text.strip())
        data.append(row_data)

    # Create a DataFrame (a data structure provided by pandas) for the scraped data
    table_data = pd.DataFrame(data, columns=headers)

    # Print the scraped data
    print(table_data)
else:
    print("Table not found in the HTML.")
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Here’s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Assuming you have already obtained the BeautifulSoup object named `soup_obj` as shown in your previous code

# Find the HTML table element by its tag name (e.g., 'table')
table = soup_obj.find('table')

# Check if the table element exists
if table:
    # Initialize lists to store table data
    ids = []
    terrestrial_dates = []
    sols = []
    ls_values = []
    months = []
    min_temperatures = []
    pressures = []

    # Extract table data from rows and cells
    data_rows = table.find_all('tr')[1:]  # Skip the first row with headers

    for row in data_rows:
        cells = row.find_all('td')

        # Check if there are enough cells in the row
        if len(cells) == 7:
            id_, terrestrial_date, sol, ls, month, min_temp, pressure = [cell.text.strip() for cell in cells]

            ids.append(id_)
            terrestrial_dates.append(terrestrial_date)
            sols.append(sol)
            ls_values.append(ls)
            months.append(month)
            min_temperatures.append(min_temp)
            pressures.append(pressure)

    # Create a DataFrame
    data = {
        'id': ids,
        'terrestrial_date': terrestrial_dates,
        'sol': sols,
        'ls': ls_values,
        'month': months,
        'min_temp': min_temperatures,
        'pressure': pressures
    }
    table_data = pd.DataFrame(data)

```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import argrelextrema

# Assuming you have a DataFrame with daily minimum temperature data
# Replace 'table_data' with your actual DataFrame
# Assuming 'sol' is the Martian day (sol) and 'min_temp' is the minimum temperature

# Create a line plot of the daily minimum temperature with the x-axis representing Martian days (sols)
plt.figure(figsize=(12, 6))
plt.plot(table_data['sol'], table_data['min_temp'], marker='o', markersize=3, linestyle='-')
plt.xlabel('Martian Days (Sols)')
plt.ylabel('Minimum Temperature (°C)')
plt.title('Daily Minimum Temperature on Mars')
plt.grid(True)

# Find local maxima and minima
minima_indices = argrelextrema(table_data['min_temp'].values, np.less, order=1)[0]
maxima_indices = argrelextrema(table_data['min_temp'].values, np.greater, order=1)[0]

# Calculate the distances between the largest and smallest peaks within a specified range
range_start = 100  # Specify the start of the range
range_end = 200    # Specify the end of the range

if len(minima_indices) > 0 and len(maxima_indices) > 0:
    minima_within_range = minima_indices[(minima_indices >= range_start) & (minima_indices <= range_end)]
    maxima_within_range = maxima_indices[(maxima_indices >= range_start) & (maxima_indices <= range_end)]
    
    if len(minima_within_range) > 0 and len(maxima_within_range) > 0:
        min_dist = np.min(np.abs(minima_within_range - maxima_within_range))
        max_dist = np.max(np.abs(minima_within_range - maxima_within_range))
        print("Minimum distance:", min_dist)
        print("Maximum distance:", max_dist)
    else:
        print("No minima and maxima within the specified range.")
else:
    print("No minima and maxima found in the data.")

# Show the plot
plt.show()
```
